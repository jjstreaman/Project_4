```python
df_processed = pd.read_csv('data/processed.csv', encoding='latin-1', index_col=0)
df_processed
```

```python

# Assuming 'df_filtered' contains your reviews and ratings
# And 'lemmatized_review' is the column with the review text

# Filter the DataFrame for negative and positive reviews
negative_reviews = df_processed[df_processed['nps_rating'] == 0]['lemmatized_review']
positive_reviews = df_processed[df_processed['nps_rating'] == 1]['lemmatized_review']

# Concatenate all reviews of each class into a single string
negative_text = " ".join(review for review in negative_reviews)
positive_text = " ".join(review for review in positive_reviews)

# Generate a word cloud image for negative reviews
wordcloud_neg = WordCloud(max_font_size=50, max_words=100, collocations=False, background_color="white").generate(negative_text)

# Generate a word cloud image for positive reviews
wordcloud_pos = WordCloud(max_font_size=50, max_words=100, collocations=False, background_color="white").generate(positive_text)

# Display the generated image:
# the matplotlib way:
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# Negative reviews word cloud
axs[0].imshow(wordcloud_neg, interpolation='bilinear')
axs[0].set_title('Word Cloud for Negative Reviews')
axs[0].axis('off')

# Positive reviews word cloud
axs[1].imshow(wordcloud_pos, interpolation='bilinear')
axs[1].set_title('Word Cloud for Positive Reviews')
axs[1].axis('off')

# Adjust layout
plt.tight_layout()

# Show the word clouds
plt.show()
```

Train Test Split

```python
X = df_processed['lemmatized_review']
```

```python
y = df_processed['nps_rating']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Test out 3 models: Logistic Regression, Random Forest, and Gradient Boosting Trees

Check best parameters for Logistic Regression

```python
vectorizers_params2 = {
    'count_vect': {
        'vectorizer': CountVectorizer(),
        'params': {
            'vect__min_df': [0.003, 0.01, 0.05]}
    },
        }
```

```python
for vect_name, vect_info in vectorizers_params2.items():
    # Set up the pipeline with the Count Vectorizer
    # try different scalers 
    logreg_pipeline = Pipeline([
        ('vect', CountVectorizer()),
        ('scale', MaxAbsScaler()),
        ('clf', LogisticRegression(max_iter=10000))
    ])
    
    # Define the parameter grid
    param_grid = vect_info['params'].copy()  # Make a copy to avoid altering the original dict
    param_grid.update({
        'clf__C': np.logspace(-4, 4, 20)  # Correct parameter for LogisticRegression
    })

    # Set up and run the GridSearchCV
    grid_search = GridSearchCV(logreg_pipeline, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train, y_train)

    # Print the best parameters and score for the current vectorizer
    print(f"Best parameters for {vect_name}: {grid_search.best_params_}")
    print(f"Best cross-validation score for {vect_name}: {grid_search.best_score_}")
```

Now, pipeline and Confusion Matrix

```python
log_reg_pipeline = Pipeline([
    ('vect', CountVectorizer(min_df=0.003)),
    ('clf', LogisticRegression(C=11.3, max_iter=10000))
])

log_reg_pipeline.fit(X_train, y_train)
```

```python
y_pred_log = log_reg_pipeline.predict(X_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_log)
```

```python
print(classification_report(y_test, y_pred_log))
```

Random Forest

```python
for vect_name, vect_info in vectorizers_params2.items():
    # Set up the pipeline with the Count Vectorizer
    # try different scalers 
    rf_pipeline = Pipeline([
        ('vect', CountVectorizer()),
        ('clf', RandomForestClassifier())
    ])
    
    # Define the parameter grid
    param_grid = vect_info['params'].copy()  # Make a copy to avoid altering the original dict
    param_grid.update({
        'clf__n_estimators': [100, 200, 300],
        'clf__max_depth': [None, 10, 20]
    })

    # Set up and run the GridSearchCV
    grid_search = GridSearchCV(rf_pipeline, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train, y_train)

    # Print the best parameters and score for the current vectorizer
    print(f"Best parameters for {vect_name}: {grid_search.best_params_}")
    print(f"Best cross-validation score for {vect_name}: {grid_search.best_score_}")
```

```python
rf_pipeline = Pipeline([
    ('vect', CountVectorizer(min_df=0.003)),
    ('clf', RandomForestClassifier(n_estimators=200, random_state=42))
])

rf_pipeline.fit(X_train, y_train)
```

```python
y_pred_rf = rf_pipeline.predict(X_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_rf)
```

```python
print(classification_report(y_test, y_pred_rf))
```

Gradient Boosting Trees

```python

for vect_name, vect_info in vectorizers_params2.items():
    # Set up the pipeline with the Count Vectorizer 
    gbt_pipeline = Pipeline([
        ('vect', CountVectorizer()),
        ('clf', GradientBoostingClassifier(random_state=42))
    ])
    
    # Define the parameter grid
    param_grid = vect_info['params'].copy()  # Make a copy to avoid altering the original dict
    param_grid.update({
        'clf__n_estimators': [100, 200],
        'clf__learning_rate': [0.01, 0.1, 0.2],
        'clf__max_depth': [3, 5, 7]
    })

    # Set up and run the GridSearchCV
    grid_search = GridSearchCV(gbt_pipeline, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train, y_train)

    # Print the best parameters and score for the current vectorizer
    print(f"Best parameters for {vect_name}: {grid_search.best_params_}")
    print(f"Best cross-validation score for {vect_name}: {grid_search.best_score_}")
```

```python
gbt_pipeline = Pipeline([
    ('vect', CountVectorizer(min_df=0.003)),
    ('clf', GradientBoostingClassifier(learning_rate=0.2, max_depth=7, n_estimators=200, random_state=42))
])

gbt_pipeline.fit(X_train, y_train)
```

```python
y_pred_gbt = gbt_pipeline.predict(X_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_gbt)
```

```python
print(classification_report(y_test, y_pred_gbt))
```

Random Forest had the highest accuracy score, and performed the best.
